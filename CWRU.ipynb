Test stand consists of a 2 hp motor (left), a torque transducer/encoder (center), a dynamometer (right), and control electronics (not shown)

## Dataset consists of:

Faults ranging from **0.007 inches in diameter to 0.040 inches** **in diameter(7,14,21,28,40)** were introduced separately at the **inner raceway, rolling element (i.e. ball) and outer raceway(3,6,12 clock angel)**. 

Faulted bearings were reinstalled into the test motor and vibration data was recorded for **motor loads of 0 to 3 horsepower (motor speeds of 1797 to 1720 RPM).**

***Two kinds: DriveEndFault & FanEndFault***

```python
class CWRUDataset(object):
  def __init__(self, data_dir, files_dic, start_idx, end_idx, skips, channels, lpf, fluctuation, batch_size):
    self.data_dir = data_dir
    self.files_dic = files_dic
    self.start_idx = start_idx
    self.end_idx = end_idx
    self.skips = skips # データの間引き間隔(1以上)
    self.channels = channels # センサチャネル數
    self.lpf = lpf # Low pass 閾値
    self.fluctuation = fluctuation # ランダム変動量
    self.batch_size = batch_size #test batch+ train bath

    # 測定データ読み取り
    self.skip_dataset, self.skip_dataset_ab = self.__get_ori_datasets()

    # 測定データ前処理 without LPF in CWRU
#    self.processed_dataset = self.__low_pass_filter(self.__norm(self.skip_dataset))
#    self.processed_dataset_ab = self.__low_pass_filter(self.__norm(self.skip_dataset_ab))
    self.processed_dataset = self.__norm(self.skip_dataset)
    self.processed_dataset_ab = self.__norm(self.skip_dataset_ab)


  def __iter__(self): 
    return self
 
  def __next__(self):  
    """学習用正常/異常データバッチ作成 shape(batch_size,time_steps,channels)
    """
    #length of each data might different,setting the length of sampling
    time_steps = int(200) 
    length = self.batch_size * time_steps
    normal_file_num = len(self.files_dic['normal_file'])
    
    rand_neg = random.uniform(0, 2 * self.fluctuation) # 一様分布乱数
    rand_pos = random.uniform(0, 2 * self.fluctuation)
    #2-dimension data matrix: (time,vibaration)
    neg_ori_data = (1 - self.fluctuation + rand_neg) * self.processed_dataset[-length:]
    pos_ori_data = (1 - self.fluctuation + rand_pos) * self.processed_dataset_ab[-length:]
    
    #training:testing = 9:1, testing_count: testing batch number; traing_count: testing batch number;
    testing_count = int(self.batch_size*0.1)
    traing_count = int(batch_size * 0.9)

    #neg_training_count: evenly read frome the normal files: trainign batch shape in every files
    neg_training_count = int(self.batch_size // normal_file_num * 0.9)

    #changing the abnormal matrix to the 2 channels:neg_data_to2
    #sampling from the same time and then concate

    neg_data_to2 = np.zeros((len(neg_ori_data), self.channels))
    for i in range(normal_file_num):
        neg_data_to2[neg_training_count*i*time_steps:neg_training_count*(i+1)*time_steps] = neg_ori_data[-(testing_count+neg_training_count)*time_steps:-testing_count*time_steps, 2*i:2*i+2]
        
    #positive and negtive batch forming 
    pos_data = pos_ori_data.reshape(-1,time_steps,self.channels)
    neg_data = neg_data_to2.reshape(-1,time_steps,self.channels)
  
    pos_train_batch = pos_data[:traing_count]
    neg_train_batch = neg_data[:traing_count]

    sample = {'pos_train_data': pos_train_batch, 
              'neg_train_data': neg_train_batch}  
    
    return sample

  def __norm(self, data):
    """正規化
    """

    return (data - data.min(axis=0)) / (data.max(axis=0) - data.min(axis=0))

  def __low_pass_filter(self, data):
    """LPFでノイズ除去
    """

    results = None
    temp = np.zeros(data.shape[0])
    for i in range(data.shape[1]):
      temp = data[:, i]
      temp_f = np.fft.fft(temp) / (len(temp) / 2)
      temp_f[self.lpf:len(temp)] = 0
      temp_if = np.fft.ifft(temp_f) * (len(temp) / 2)
      if results is None:
        results = temp_if.reshape(-1, 1)
      else:
        results = np.hstack((results, temp_if.reshape(-1, 1)))
    
    return results

  #normal files & abnormal file divide into different matrix
  def __get_ori_datasets(self): 
    """測定データ読み取り
    """
    #abnormal file--abnomal matrix 2 channels
    #ab_filename: abnormal file; ab_data: abnormal data matrix
    ab_filename = self.files_dic['abnormal_file']
    ab_data = np.loadtxt(os.path.join(data_dir, ab_filename), delimiter=' ',skiprows=1,usecols=(0,1)) #only use the FE&DE column
    ab_data = ab_data[self.start_idx:self.end_idx, :].reshape(-1, 2)

    #normal files--normal matrix: files_num*2channels
    #nomal_data: the final normal matrix; norm_data: every normal file matrix
    normal_data = None
    for filename in tqdm(self.files_dic['normal_file'], desc='Loading CWRU files.'):
      norm_data = np.loadtxt(os.path.join(data_dir, filename), delimiter=' ',skiprows=1,usecols=(0,1))#only use the FE&DE column
      if normal_data is None:
        normal_data = norm_data[self.start_idx:self.end_idx, :].reshape(-1, 2)
      else:
        normal_data = np.hstack((normal_data,norm_data[self.start_idx:self.end_idx, :]))

    #skips sampling and transfer the abnormal&normal matrix to form the nwe batch 
    #skip_dataset: normal matrix after skip sampling; skip_dataset_ab: abnormal matrix after skip sampling
    steps, steps_ab = len(normal_data), len(ab_data)
    skip_dataset = None
    skip_dataset_ab = None
    #nornal dataset:every files consists of 2 channels in matrix
    #normal_channels: the sum of all channels
    normal_channels = normal_data.shape[1]
    for i in tqdm(range(int(steps/self.skips)), desc='Processing normal data.'):
      if skip_dataset is None:
        skip_dataset = normal_data[i*self.skips, :normal_channels].reshape(-1, normal_channels)
      else:
        skip_dataset = np.vstack((skip_dataset, normal_data[i*self.skips, :normal_channels]))

    for i in tqdm(range(int(steps_ab/self.skips)), desc='Processing abnormal data.'):
      if skip_dataset_ab is None:
        skip_dataset_ab = ab_data[i*self.skips, :self.channels]
      else:
        skip_dataset_ab = np.vstack((skip_dataset_ab, ab_data[i*self.skips, :self.channels]))

    return skip_dataset, skip_dataset_ab

  def plot_data(self):
    for i in range(len(self.files_dic['normal_file'])):
      plt.rcParams['figure.figsize'] = (20, 4.8)
      plt.rcParams['font.size'] = 18
      plt.subplot(xlabel='Time (sec)', ylabel='Normal Vibaration (degree)')
      plt.plot(self.skip_dataset[0:200,2*i:2*i+2])
      plt.show()

    plt.rcParams['figure.figsize'] = (20, 8)
    plt.rcParams['font.size'] = 18
    plt.subplot(xlabel='Time (sec)', ylabel='Abnormal Vibaration  (degree)')
    plt.plot(self.skip_dataset_ab[0:200])
    plt.show()

    for i in range(len(self.files_dic['normal_file'])):
      plt.rcParams['figure.figsize'] = (20, 4.8)
      plt.rcParams['font.size'] = 18
      plt.subplot(xlabel='Time (sec)', ylabel='Normal Normalized Vibaration (degree)')
      plt.plot(self.processed_dataset[0:200,2*i:2*i+2])
      plt.show()

    plt.rcParams['figure.figsize'] = (20, 8)
    plt.rcParams['font.size'] = 18
    plt.subplot(xlabel='Time (*10 sec)', ylabel='Abnormal Normalized Vibaration  (degree)')
    plt.plot(self.processed_dataset_ab[0:200])
    plt.show()

  def get_data(self):
    return self.processed_dataset, self.processed_dataset_ab

  def get_test_batch(self):
    """テスト用バッチ作成
       all test batch coming from the last period of the original data
       test batch shape:(test_batch_size,time_steps,channels) = (self.batch_size*test proportion,time_steps,self.channels)
       abnormal test batch: from the one abnormal file, abnormal test batch size = self.batch_size*test proportion
       normal test batch: forming from every normal file, batch size = self.batch_size*test proportion*normal_file_num
    """
    normal_file_num = len(self.files_dic['normal_file'])
    time_steps = int(200) 
    testing_count = int(self.batch_size*0.1)

    #neg_data_to2: transfer the original abnormal matrix(normal_file_num*2channels) to the expected 2-channels shape
    #in one testing count batch: coming from one abnomal files
    #then arrange then as the order of original abnormal matrix
    neg_data_to2 = None
    for i in tqdm(range(normal_file_num), desc='Processing normal testing batch data.'):
      if neg_data_to2 is None:
        neg_data_to2 = self.skip_dataset[-testing_count*time_steps:, 0:2].reshape(-1, 2) 
      else:
        neg_data_to2 = np.hstack((neg_data_to2,self.skip_dataset[-testing_count*time_steps:, 2*i:2*i+2].reshape(-1,2)))

    #pos_batch: positive testing batch, dividing  from the end abnormal matrix
    #neg_batch: negative testing batch, reading from the former neg_data_to2 matrix
    pos_batch = self.skip_dataset_ab[-testing_count*time_steps:].reshape(-1,time_steps,self.channels)
    neg_batch = neg_data_to2.reshape(-1,time_steps,self.channels)   

    sample={'pos_test_data': pos_batch, 
              'neg_test_data': neg_batch}
    return sample
```

```python
class Cov1DModel(tf.keras.Model):
    def __init__(self):
        he_initializer = tf.keras.initializers.he_normal()
        norm_initializer = tf.keras.initializers.RandomNormal(mean=1.0, stddev=0.01)
        super(TtsModel, self).__init__()
        self.conv1d1 = layers.Conv1D(128, kernel_size=(9,), kernel_initializer=he_initializer, padding='same', activation='relu', input_shape=(40,200,12))
        self.norm1 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
        self.conv1d2 = layers.Conv1D(128, kernel_size=(9,), kernel_initializer=he_initializer, padding='same', activation='relu')
        self.norm2 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
        self.conv1d3 = layers.Conv1D(1, kernel_size=(1,), kernel_initializer=he_initializer, activation='sigmoid')
        self.pooling = layers.AveragePooling1D(pool_size=10)

    def call(self, x):
        x = self.conv1d1(x)
        x = self.norm1(x)
        x = self.conv1d2(x)
        x = self.norm2(x)
        x = self.conv1d3(x)    
        x = self.pooling(x)
        return x
```

```python
# Masking 3 dimesion
#re-marking 3rd dimension is the real datasets 
#reshape to the same shape
#1 - input padding masking : Lower triangular part hidden mask
def generate_mask(input):
    input  = tf.cast(tf.math.equal(input,0),tf.float32)
    return tf.transpose(input[:,:, tf.newaxis,:],perm=[0,3,2,1])
    
#from other repo mask https://github.com/bryanlimy/tf2-transformer-chatbot/blob/master/transformer/model.py  
def create_padding_mask(x):
  mask = tf.cast(tf.math.equal(x, 0), dtype=tf.float32)
  return mask[:, : , tf.newaxis, : ]

#2 mask the sequence behind the predicting sequence, need to compare to the padding mask since decoding also need the padding mask
def generate_lookahead_mask(input):
    seq = tf.shape(input)[1]
    paddings = tf.constant([[10, 0,], [0, 10]])
    down_linalg = tf.pad(tf.linalg.band_part(
        tf.ones((seq-10, seq-10), dtype=tf.float32), -1, 0),paddings,"CONSTANT")
    mask = 1 - tf.linalg.band_part(tf.ones((seq,seq)),-1,0)
    mask = mask + down_linalg

    return mask
    
#from other repo mask : decoder mask, combine two part, assume the input as 2D
def create_look_ahead_mask(x):
    seq_len = tf.shape(x)[1]
    look_ahead_mask = 1 - tf.linalg.band_part(
        tf.ones((seq_len, seq_len), dtype=tf.float32), -1, 0)
    padding_mask = create_padding_mask(x)
    return tf.maximum(look_ahead_mask, padding_mask)
```

```python
def self_attention(q,k,v,mask,d_model): #d_model vector_size
    qk = tf.matmul(q,k,transpose_b = True) #(batch*seq_len*seq_len)
    scaled_qk = tf.divide(qk,tf.math.sqrt(tf.cast(d_model,tf.float32)))

    if mask is not None:
        scaled_qk += (mask*-1e9)
    attention_weights = tf.nn.softmax(scaled_qk,axis = -1)
    final_output = tf.matmul(attention_weights,v) #(batch*seq_len*vocab_Size)

    return attention_weights,final_output
```

```python
class MultiHeadAttention(tf.keras.layers.Layer): #vector size = d_model--third dimension
    def __init__(self,num_heads,d_model):
        super(MultiHeadAttention,self).__init__()
        self.num_heads = num_heads
        self.d_model = d_model

        assert self.d_model % self.num_heads ==0 

        self.depth = self.d_model//self.num_heads
        #dense layer
        self.wq = tf.keras.layers.Dense(units = d_model)
        self.wk = tf.keras.layers.Dense(units = d_model)
        self.wv = tf.keras.layers.Dense(units = d_model)
        self.dense1 = tf.keras.layers.Dense(units = d_model)

#input--input data generate weight and then divide into multi-heads part for calculate different attention 
    def split_into_heads(self,input,batch_size):

        input = tf.reshape(input,(batch_size,-1,self.num_heads,self.depth))

        return tf.transpose(input,perm = [0,2,1,3]) # batch size, heads, len, vector size
    
    def build(self, input_shape):
        self.q = self.add_weight(
            shape=(input_shape[-1], self.d_model),initializer="random_normal",trainable=True,
        )
        self.k = self.add_weight(
            shape=(input_shape[-1], self.d_model), initializer="random_normal", trainable=True
        )
        self.v = self.add_weight(
            shape=(input_shape[-1], self.d_model), initializer="random_normal", trainable=True
        )   
    
    def call(self,input,mask):
        batch_size = input.shape[0]

        #linear layer to generate the weight matrix
        q = self.wq(input)
        k = self.wk(input)
        v = self.wq(input)
        #splitting into heads
        q = self.split_into_heads(q,batch_size)
        k = self.split_into_heads(k,batch_size)
        v = self.split_into_heads(v,batch_size)

        attention_weights,multi_head_output = self_attention(q,k,v,mask,self.depth) #for every head
        pre_concat_multi_head_output =  tf.transpose(multi_head_output,perm = [0,2,1,3])
        
        post_concat_multi_head_output =  tf.reshape(pre_concat_multi_head_output,(batch_size,-1,self.d_model))
        
        final_output = self.dense1(post_concat_multi_head_output)

        return attention_weights,final_output
```

```python
class MultiHeadAttention2(tf.keras.layers.Layer):
    def __init__(self,num_heads,d_model):
        super(MultiHeadAttention2,self).__init__()
        self.num_heads = num_heads
        self.d_model = d_model

        assert self.d_model % self.num_heads ==0 

        self.depth = self.d_model//self.num_heads

        self.wq = tf.keras.layers.Dense(units = d_model)
        self.wk = tf.keras.layers.Dense(units = d_model)
        self.wv = tf.keras.layers.Dense(units = d_model)
        self.dense1 = tf.keras.layers.Dense(units = d_model)

#x--input series divide into heads part, add the dimension of the turn of the head
    def split_into_heads(self,x,batch_size):
        #print("x shape is",x.get_shape())
        x = tf.reshape(x,(batch_size,-1,self.num_heads,self.depth))
#        print("x shape is",x.get_shape())
        return tf.transpose(x,perm = [0,2,1,3])
     
    def build(self, input_shape):
        self.q = self.add_weight(
            shape=(input_shape[-1], self.d_model),initializer="random_normal",trainable=True,
        )
        self.k = self.add_weight(
            shape=(input_shape[-1], self.d_model), initializer="random_normal", trainable=True
        )
        self.v = self.add_weight(
            shape=(input_shape[-1], self.d_model), initializer="random_normal", trainable=True
        )   
    def call(self,enc_output,dec_input,mask):
        batch_size = enc_output.shape[0]
        
        #linear layer 
        self.q = self.wq(dec_input)
        self.k = self.wk(enc_output)
        self.v = self.wq(enc_output)

        #splitting into heads
        q = self.split_into_heads(q,batch_size)
        k = self.split_into_heads(k,batch_size)
        v = self.split_into_heads(v,batch_size)

        attention_weights,multi_head_output = self_attention(q,k,v,mask,self.depth) #for every head

        pre_concat_multi_head_output =  tf.transpose(multi_head_output,perm = [0,2,1,3])
        post_concat_multi_head_output =  tf.reshape(pre_concat_multi_head_output,(batch_size,-1,self.d_model))
        
        final_output = self.dense1(post_concat_multi_head_output)

        return attention_weights,final_output
```

```python
#embedding --similarity, without the positional distance information. Thus add the position on embeddings
class get_positional_encoding(tf.keras.layers.Layer):
  def __init__(self,position,d_model):
    super(get_positional_encoding,self).__init__()
    self.position = position
    self.d_model = d_model
    self.pos_encoding = self.positional_encoding(position,d_model)
  
  def get_angle(self,position,i,d_model):
    angle = 1 / tf.pow(10000, (2*(i//2))/ tf.cast(d_model,tf.float32))
    return position * angle

  def positional_encoding(self, position, d_model): #d_model = 512 if 128
    angle_final = self.get_angle(tf.expand_dims(tf.range(tf.cast(position,tf.float32)),axis = 1),tf.expand_dims(tf.range(tf.cast(d_model,tf.float32)),axis = 0),d_model)
    np_angle_final = angle_final.numpy()
    np_angle_final[:,0::2] = np.sin(np_angle_final[:,0::2])
    np_angle_final[:,1::2] = np.cos(np_angle_final[:,1::2])
    angle_final = tf.convert_to_tensor(np_angle_final)
    pos_encoding = angle_final[tf.newaxis,...]
    return tf.cast(pos_encoding,dtype=tf.float32)
  
  def call(self,input):
    return input + self.pos_encoding[:, :tf.shape(input)[1], : ]
```

```python
#hidden_layer_shape--units(how many to enlarge) generally, d_model*4 
class FeedForward_Layer(tf.keras.layers.Layer): 
  def __init__(self,d_model,hidden_layer_shape):
        super(FeedForward_Layer,self).__init__()
        self.d_model = d_model
        self.hidden_layer_shape = hidden_layer_shape
        self.dense1 = tf.keras.layers.Dense(hidden_layer_shape, activation='relu') # (batch_size, seq_len, dff) relu-->gelu
        self.dense2 = tf.keras.layers.Dense(d_model)  # (batch_size, seq_len, d_model) normalization done on the final d_model axis

  def call(self,input):
        output = self.dense1(input)
        final_output = self.dense2(output)
        return final_output
  
```

```python
class Encoding_Layer(tf.keras.layers.Layer):
    def __init__(self,num_heads,d_model,hidden_layer_shape,rate=0.1):
        super(Encoding_Layer,self).__init__()
        self.d_model = d_model
        self.num_heads = num_heads
        self.multi_head_attention = MultiHeadAttention(num_heads,d_model)
        self.ffd_net = FeedForward_Layer(d_model,hidden_layer_shape)
        self.norm1 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.norm2 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.dropout1 = tf.keras.layers.Dropout(rate)
        self.dropout2 = tf.keras.layers.Dropout(rate)
    
    def call(self,input,mask,training = True):
        #part 1
        _,multihead_output = self.multi_head_attention(input,mask)
        dropout_multihead_output = self.dropout1(multihead_output,training = training)
        skip_dropout_output = tf.add(input,dropout_multihead_output)
        norm1_skip_output  = self.norm1(skip_dropout_output)        
        #part 2
        point_norm1_output = self.ffd_net(norm1_skip_output)
        dropout_point_output = self.dropout2(point_norm1_output,training = training)
        skip_dropout_output = tf.add(dropout_point_output,norm1_skip_output)
        norm2_skip_output = self.norm2(skip_dropout_output)
        return norm2_skip_output
```

```python
class Decoding_Layer(tf.keras.layers.Layer):
    def __init__(self,num_heads,d_model,hidden_layer_shape,rate = 0.1):
        super(Decoding_Layer,self).__init__()
        self.multi_head_attention1 = MultiHeadAttention(num_heads,d_model)
        self.multi_head_attention2 = MultiHeadAttention2(num_heads,d_model)
        self.norm1 = tf.keras.layers.LayerNormalization(epsilon = 1e-6) #mean=1 variance=0
        self.norm2 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.norm3 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.point_ffd_net = FeedForward_Layer(d_model,hidden_layer_shape)
        self.dropout1 = tf.keras.layers.Dropout(rate)
        self.dropout2 = tf.keras.layers.Dropout(rate)
        self.dropout3 = tf.keras.layers.Dropout(rate)

    def call(self,enc_output,dec_input,look_ahead_mask,padding_mask,training = True):
        #decoder attention with look ahead mask(include padding mask)
        # enc_output.shape --> (batch_size, input_seq_len, d_model)
        _,multihead1_output = self.multi_head_attention1(dec_input,look_ahead_mask)
        dropout1_multihead1_output = self.dropout1(multihead1_output)
        skip1_dropout1_output = tf.add(dropout1_multihead1_output,dec_input)
        norm1_skip1_output = self.norm1(skip1_dropout1_output)

        #decoder&encoder attention with padding mask
        _,multihead2_output  = self.multi_head_attention2(enc_output,dec_input,padding_mask)
        dropout2_multihead2_output = self.dropout2(multihead2_output)
        skip2_dropout2_output = tf.add(dropout2_multihead2_output,norm1_skip1_output)
        norm2_skip2_output = self.norm2(skip2_dropout2_output)

        #sum and normalize
        point_norm2_output = self.point_ffd_net(norm2_skip2_output)
        dropout3_point_output = self.dropout3(point_norm2_output)
        skip3_dropout3_output = tf.add(norm2_skip2_output,dropout3_point_output)
        norm3_skip3_output = self.norm3(skip3_dropout3_output)
        return norm3_skip3_output
```

```python
#without embeddings
class Encoder(tf.keras.Model):
    def __init__(self,d_model,num_heads,hidden_layer_shape,max_pos_encoding,num_enc_layers,rate = 0.1):
        super(Encoder,self).__init__()
        self.max_pos_encoding = max_pos_encoding #max_pos_encoding--longest position 
        self.num_enc_layers = num_enc_layers
        self.d_model = d_model
        self.num_heads = num_heads
        self.hidden_layer_shape = hidden_layer_shape
        self.position_encoding = get_positional_encoding(max_pos_encoding,d_model)#using part of the position
        self.encoder_layers = [Encoding_Layer(num_heads,d_model,hidden_layer_shape,rate) for _ in range(num_enc_layers)]
        self.dropout = tf.keras.layers.Dropout(rate)

    def call(self,inputs,training):
        enc_padding_mask = None
        pos_scaled_embedding = tf.math.add(inputs,self.position_encoding(inputs))
        output_input_encoder = self.dropout(pos_scaled_embedding,training = training)

        for i in range(self.num_enc_layers):
            output_input_encoder = self.encoder_layers[i](output_input_encoder,enc_padding_mask,training)

        return output_input_encoder
```

```python
#without embeddings
class Decoder(tf.keras.Model):
    def __init__(self,d_model,num_heads,hidden_layer_shape,max_position_encoding,num_layers,rate=0.1): #dec_input = decoder input
        super(Decoder,self).__init__()
        self.position_encoding = get_positional_encoding(max_position_encoding,d_model)
        self.d_model = d_model
        self.num_layers = num_layers
        self.decoder_layers = [Decoding_Layer(num_heads,d_model,hidden_layer_shape) for _ in tf.range(num_layers)]
        self.dropout = tf.keras.layers.Dropout(rate)
        self.num_layers = num_layers

    def call(self,input,training = True):
        #positional_encoding = self.position_encoding[:,:seq_len,:]
        dec_input=input[0]
        enc_output = input[1]
        look_ahead_mask = generate_lookahead_mask(dec_input)
#        dec_padding_mask = create_padding_mask(dec_input)
        padding_mask = None
        positional_encoding = self.position_encoding(dec_input)
        pos_scaled_embedding = tf.math.add(dec_input,positional_encoding)
        output_input_decoder = self.dropout(pos_scaled_embedding,training = training)

        for i in range(self.num_layers):
            output_input_decoder = self.decoder_layers[i](output_input_decoder,enc_output,look_ahead_mask,padding_mask)
        
        return output_input_decoder
```

```python
#With decoder
class Transformer(tf.keras.Model):
    def __init__(self,num_layers,num_heads,d_model,hidden_layer_shape,max_pos_encoding, rate = 0.1):
        super(Transformer,self).__init__()
        self.encoder = Encoder(d_model,num_heads,hidden_layer_shape,max_pos_encoding,num_layers,rate) #pe_input--max_pos_encoding
        self.decoder = Decoder(d_model,num_heads,hidden_layer_shape,max_pos_encoding,num_layers,rate) #pe_output--max_pos_encoding dec_input_Size--detect input size
        self.pooling = layers.AveragePooling1D(pool_size=10)
        self.final_layer = tf.keras.layers.Dense(1)

    def call(self,input,training = True):
        
        #mid = int(input.shape[0])
        enc_input=input[0]
        dec_input=input[1]
        enc_output = self.encoder(enc_input,training)
#        final_dec_input = tf.concat(dec_input,enc_output,0)
        final_dec_input = [dec_input,enc_output]
        dec_output = self.pooling(self.decoder(final_dec_input,training))
        final_output = self.final_layer(dec_output)

        return final_output
```

# **3. Training:**

```python
#training batch = 45
max_step = 500 # 学習の繰り返し回数
m1 = 0.001 #正規化項内の定数
m2 = 0.001
loss_value =[]

process = tqdm(enumerate(CWRUdataset), total=max_step)
for step, data in process:
  if step == max_step:
    break
  pos_data, neg_data = data['pos_train_data'], data['neg_train_data']
  batch = tf.concat([pos_data, neg_data],0)
  #[90,200,2]のbatchが出来上がる

  loss = 0.0 #stepごとにlossを初期化
  with tf.GradientTape() as t:
      sum1 = 0.0
      sum2 = 0.0 
      result = tf.zeros([90, 0, 1])
      result = tf.cast(result,tf.float32)
      #result[0]からresult[44]が異常データに対する推定値, result[45]からresult[89]が正常データに対する推定値になっている
      for i in range(0, 200, 10):
          buf = model(batch[:, i:i+10])
          result = tf.concat([result, buf], axis=1)
#          result = tf.cast(result,tf.int32)
      l_sum = []
      for i in range(45):
          y_pos_max = tf.math.reduce_max(result[i])
          y_neg_max = tf.math.reduce_max(result[i+45])
          l = tf.math.reduce_max([0, tf.math.add(1.0, tf.math.subtract(y_neg_max, y_pos_max))]) #波形1セット分の損失
          l_sum.append(l)

      loss = tf.math.reduce_sum(l_sum)
      
      #正規化項を計算
      pos_result = np.squeeze(result[0:45], axis=2).flatten()
      neg_result = np.squeeze(result[45:], axis=2).flatten()
      number = []
      for k in range(len(pos_result)-1):
          number.append(tf.square(tf.math.subtract(pos_result[k], pos_result[k+1])))
      sum1 = tf.math.reduce_sum(number)
      sum2 = tf.math.reduce_sum(pos_result)
      smooth = tf.math.scalar_mul(m1, sum1)
      sparse = tf.math.scalar_mul(m2, sum2)
      lasso = tf.math.add(smooth, sparse)
      loss = tf.math.add(loss, lasso)
      loss_value.append(loss)

  d = t.gradient(loss, model.trainable_weights)
  optimizer.apply_gradients(zip(d, model.trainable_weights))
  print(loss_value[-1])
  postfix = 'Step: {0:4d}, Loss: {1:.3f}'.format(step+1, loss)
  process.set_postfix_str(postfix)
```
