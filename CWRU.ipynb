Faulted bearings were reinstalled into the test motor and vibration data was recorded for **motor loads of 0 to 3 horsepower (motor speeds of 1797 to 1720 RPM).**

```python
class CWRU_CVDataset(object):
  def __init__(self, data_dir,output_dir, files_dic, file_list, split, channels, time_steps, batch_size, dividing):
    self.data_dir = data_dir #data_files
    self.files_dic = files_dic #abnormal and noraml files index
    self.file_list = file_list #all file list
    self.output_dir = output_dir #sampled data saving
    self.channels = channels #channels used for training
    self.batch_size = batch_size #postive batch = negative batch = batch_size (including taing and validation)
    self.time_steps = time_steps # time_stpes in every sample
    self.split = split #k-fold cross validation, split = k
    self.dividing = dividing #control whether divide the dataset or not. True->divide False->not divide
    if self.dividing is True:#for the same data_list, can divide it only once
        self.divide_sample = self.divide_sample() #dividing the oringinal files to time_steps sample like
    self.select = self.select() #randomly select the file from divided samples
    self.dataset_ab,self.dataset= self.loading_data() #loading the sampled data as the index like
    self.test_batch_ab,self.test_batch = self.get_test_batch()#forming the test batch
    self.processed_dataset = self.__norm(self.dataset) #normalization the dataset
    self.processed_dataset_ab = self.__norm(self.dataset_ab)
    self.processed_test_batch = self.__norm(self.test_batch) 
    self.processed_test_batch_ab = self.__norm(self.test_batch_ab) 

  
  def divide_sample(self):
    '''divide the original dataset into sample
    '''
    for filename in tqdm(self.file_list, desc='Dividing CWRU files.'):
      data = np.loadtxt(os.path.join(self.data_dir, filename), delimiter=' ',skiprows=1,usecols=(0,1)) #only use the FE&DE column
      sample_num = len(data)//self.time_steps #sample_num: the number of the sampling for each file
      index = random.sample(range(0,sample_num),int(self.batch_size*1.1)) #randomly select samples. 
      #the number of selection: training batch size + validation batch size + test batch size = 9:1:1
      #self.batch_size = traing batch + validation batch
      #assume at the same series of data, the different situations only different in the way loading the data
      out_dir = os.path.join(self.output_dir,filename) #saving in the new filefold named as its filename
      self.check_file_dir(out_dir) # delete the last time division and create for new samples
        
      for num in index:
        new_filename = filename[:-4] + '_' + str(num).rjust(4,'0')+'.txt' #rename the sample file with its order in the orginal file
        sample = data[num*self.time_steps:(num+1)*self.time_steps]
        np.savetxt(os.path.join(out_dir,new_filename), sample,fmt='%f',delimiter=' ')
    return  self


  def select(self):
    '''acording to the abnormal or normal the files belonged to, make the selection for k-fold cross validation

       abnormal files all coming from the one abnormal file:
            (the number of selection for abnormal file: batch_size)

       normal files will be selected coming from all normal files evenly 
            (the number of selection for each normal file: batch_size/the_number_of_normal_files)
       
      after selection, every fold should contain the two sub_fold: normal and subnormal
    '''
    #1.generate the new file_dir for K-Fold Validation and test filefold
    for i in range(self.split):
        self.check_file_dir(os.path.join(self.output_dir,str(i)))
    self.check_file_dir(os.path.join(self.output_dir,'test'))
    
    #2.randomly select the samples using index from the divided sample
    #2.1 normal samples select
    normal_count = self.batch_size//len(self.files_dic['normal_file']) #evenly choose, thus the nomal batch number of each is divided by normal file number
    for filename in tqdm(self.files_dic['normal_file'], desc='Selecting normal CWRU files.'):
      input_dir = os.path.join(self.output_dir,filename) #input_dir:the original samples file fold
      kth = 0 #kth: in K-Fold validation, refers to kth fold
      #count: the number of samples has be choosen,
      #then distributed in each kth fold and test file
      for count,file in enumerate(os.listdir(input_dir)):     
        #distribute the selected samples to the K folds and test
        to_path = os.path.join(self.output_dir,str(kth),'normal_file')#to path: the selected normal files copy to
        #distributed into Kth folds
        #normal file will be selected from the abnormal file list evenly
        #then distributed to K folds evenly
        if count >= normal_count//self.split*(kth+1) and count<normal_count: 
            kth += 1
            to_path = os.path.join(self.output_dir,str(kth),'normal_file')
        #distributed into test fold
        elif count < int(normal_count+batch_size*0.1):
            to_path = os.path.join(self.output_dir,'test','normal_file')
            
        elif count> int(normal_count+batch_size*0.1) :
            break
            
        if not os.path.isdir(to_path):
            os.makedirs(to_path)
        #copy the file to the to_path     
        copy(os.path.join(input_dir,file),os.path.join(to_path,file))
        
    #2.2 abnormal samples select
    filename = self.files_dic['abnormal_file'] 
    input_dir = os.path.join(self.output_dir,filename) #input_dir: the original abnormal samples location

    kth = 0 #kth: in K-Fold validation, refers to kth fold
    for count,file in enumerate(os.listdir(input_dir)):
        #distribute the selected samples into the K folds and test fold
        to_path = os.path.join(self.output_dir,str(kth),'abnormal_file')  #to path: the selected abnormal files copy to where
        #to Kth-Fold
        if count >= batch_size//self.split*(kth+1) and count<batch_size: 
            kth += 1
            to_path = os.path.join(self.output_dir,str(kth),'abnormal_file')
        #test fold
        elif count >= batch_size and count <= int(batch_size*1.1):
            to_path = os.path.join(self.output_dir,'test','abnormal_file')
            
        elif count > int(batch_size*1.1):
            break
            
        if not os.path.isdir(to_path):
            os.makedirs(to_path)
            
        copy(os.path.join(input_dir,file),os.path.join(to_path,file))

    return self
    

  def check_file_dir(self,dir):
    ''' dir: the path need to be checked 
        if the dir not existed --> make new dir
        if the dir existed and already have files --> delete the last time result
    '''
    if not os.path.isdir(dir):
        os.makedirs(dir)
    if os.listdir(dir): 
        shutil.rmtree(dir)
        os.makedirs(dir)
  

  def loading_data(self):
    ''' loading dataset from selected samples for training and validation
        finally return the postive batch--ab_data and negative batch--norm_data
    '''
    ab_data = None #ab_data: abnormal data
    norm_data = None #norm_data: normal data
    for i in range(self.split): # in every kth fold 
        input_dir = os.path.join(self.output_dir,str(i)) 
        for filefold in os.listdir(input_dir): # there are two file fold in each kth fold : abnomal_file and normal_file 
            file_dir = os.path.join(input_dir,filefold)
            #loading data from abnormal_file to form the norm_data
            if filefold == 'abnormal_file': 
                for file in os.listdir(file_dir):
                    sample = np.loadtxt(os.path.join(file_dir, file), delimiter=' ') #every selected samples
                    if ab_data is None:
                        ab_data = sample
                    else:
                        ab_data = np.vstack((ab_data,sample))
        #loading data from abnormal_file to form the ab_data
            else:
                for file in os.listdir(file_dir):
                    sample = np.loadtxt(os.path.join(file_dir, file), delimiter=' ')
                    if norm_data is None:
                        norm_data = sample
                    else:
                        norm_data = np.vstack((norm_data,sample))

    return ab_data.reshape(-1,self.time_steps,self.channels),norm_data.reshape(-1,self.time_steps,self.channels)  

  def __norm(self, data):
    '''normalization
    '''
    return (data - data.min(axis=0)) / (data.max(axis=0) - data.min(axis=0))


  def get_test_batch(self):
    ''' loading dataset from selected test samples for testing
        finally return the postive batch--ab_data and negative batch--norm_data
        for every file select the batch_size*0.1 samples
        thus the neg_batch:pos_batch = normal_files: abnormal_file = 5:1
    '''        
    ab_data = None #ab_data: abnormal data
    norm_data = None #norm_data: normal data
    #loading data from abnormal_test_file to form the ab_data
    file_dir = os.path.join(self.output_dir,'test','abnormal_file')
    for file in os.listdir(file_dir):
        sample = np.loadtxt(os.path.join(file_dir, file), delimiter=' ') #every selected samples
        if ab_data is None:
            ab_data = sample
        else:
            ab_data = np.vstack((ab_data,sample))
        
    #loading data from normal_test_file to form the norm_data
    file_dir = os.path.join(self.output_dir,'test','normal_file')
    norm_file_list = os.listdir(file_dir)
    norm_file_list.sort()
    for file in norm_file_list:
        sample = np.loadtxt(os.path.join(file_dir, file), delimiter=' ')
        if norm_data is None:
            norm_data = sample
        else:
            norm_data = np.vstack((norm_data,sample))

        
    return ab_data.reshape(-1,self.time_steps,self.channels),norm_data.reshape(-1,self.time_steps,self.channels)     
    
```

```python
class Cov1DModel(tf.keras.Model):
    def __init__(self):
        he_initializer = tf.keras.initializers.he_normal()
        norm_initializer = tf.keras.initializers.RandomNormal(mean=1.0, stddev=0.01)
        super(Cov1DModel, self).__init__()
        self.conv1d1 = layers.Conv1D(128, kernel_size=(9,), kernel_initializer=he_initializer, padding='same', activation='relu', input_shape=(40,200,12))
        self.norm1 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
        self.conv1d2 = layers.Conv1D(128, kernel_size=(9,), kernel_initializer=he_initializer, padding='same', activation='relu')
        self.norm2 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
        self.conv1d3 = layers.Conv1D(1, kernel_size=(1,), kernel_initializer=he_initializer, activation='sigmoid')
        self.pooling = layers.AveragePooling1D(pool_size=10)

    def call(self, x):
        x = self.conv1d1(x)
        x = self.norm1(x)
        x = self.conv1d2(x)
        x = self.norm2(x)
        x = self.conv1d3(x)    
        x = self.pooling(x)
        return x
```

```python
# Masking 3 dimesion
#re-marking 3rd dimension is the real datasets 
#reshape to the same shape
#1 - input padding masking : Lower triangular part hidden mask
def generate_mask(input):
    input  = tf.cast(tf.math.equal(input,0),tf.float32)
    return tf.transpose(input[:,:, tf.newaxis,:],perm=[0,3,2,1])
    
#from other repo mask https://github.com/bryanlimy/tf2-transformer-chatbot/blob/master/transformer/model.py  
def create_padding_mask(x):
  mask = tf.cast(tf.math.equal(x, 0), dtype=tf.float32)
  return mask[:, : , tf.newaxis, : ]

#2 mask the sequence behind the predicting sequence, need to compare to the padding mask since decoding also need the padding mask
def generate_lookahead_mask(input):
    seq = tf.shape(input)[1]
    paddings = tf.constant([[10, 0,], [0, 10]])
    down_linalg = tf.pad(tf.linalg.band_part(
        tf.ones((seq-10, seq-10), dtype=tf.float32), -1, 0),paddings,"CONSTANT")
    mask = 1 - tf.linalg.band_part(tf.ones((seq,seq)),-1,0)
    mask = mask + down_linalg

    return mask
    
#from other repo mask : decoder mask, combine two part, assume the input as 2D
def create_look_ahead_mask(x):
    seq_len = tf.shape(x)[1]
    look_ahead_mask = 1 - tf.linalg.band_part(
        tf.ones((seq_len, seq_len), dtype=tf.float32), -1, 0)
    padding_mask = create_padding_mask(x)
    return tf.maximum(look_ahead_mask, padding_mask)
```

```python
def self_attention(q,k,v,mask,d_model): #d_model vector_size
    qk = tf.matmul(q,k,transpose_b = True) #(batch*seq_len*seq_len)
    scaled_qk = tf.divide(qk,tf.math.sqrt(tf.cast(d_model,tf.float32)))

    if mask is not None:
        scaled_qk += (mask*-1e9)
    attention_weights = tf.nn.softmax(scaled_qk,axis = -1)
    final_output = tf.matmul(attention_weights,v) #(batch*seq_len*vocab_Size)

    return attention_weights,final_output
```

```python
class MultiHeadAttention(tf.keras.layers.Layer): #vector size = d_model--third dimension
    def __init__(self,num_heads,d_model):
        super(MultiHeadAttention,self).__init__()
        self.num_heads = num_heads
        self.d_model = d_model

        assert self.d_model % self.num_heads ==0 

        self.depth = self.d_model//self.num_heads
        #dense layer
        self.wq = tf.keras.layers.Dense(units = d_model,trainable = True)
        self.wk = tf.keras.layers.Dense(units = d_model,trainable = True)
        self.wv = tf.keras.layers.Dense(units = d_model,trainable = True)
        self.dense1 = tf.keras.layers.Dense(units = d_model,trainable = True)

#input--input data generate weight and then divide into multi-heads part for calculate different attention 
    def split_into_heads(self,input,batch_size):

        input = tf.reshape(input,(batch_size,-1,self.num_heads,self.depth))

        return tf.transpose(input,perm = [0,2,1,3]) # batch size, heads, len, vector size
    
    def call(self,input,mask):
        batch_size = input.shape[0]

        #linear layer to generate the weight matrix
        q = self.wq(input)
        k = self.wk(input)
        v = self.wq(input)
        #splitting into heads
        q = self.split_into_heads(q,batch_size)
        k = self.split_into_heads(k,batch_size)
        v = self.split_into_heads(v,batch_size)

        attention_weights,multi_head_output = self_attention(q,k,v,mask,self.depth) #for every head
        pre_concat_multi_head_output =  tf.transpose(multi_head_output,perm = [0,2,1,3])
        
        post_concat_multi_head_output =  tf.reshape(pre_concat_multi_head_output,(batch_size,-1,self.d_model))
        
        final_output = self.dense1(post_concat_multi_head_output)

        return attention_weights,final_output
```

```python
class MultiHeadAttention2(tf.keras.layers.Layer):
    def __init__(self,num_heads,d_model):
        super(MultiHeadAttention2,self).__init__()
        self.num_heads = num_heads
        self.d_model = d_model

        assert self.d_model % self.num_heads ==0 

        self.depth = self.d_model//self.num_heads

        self.wq = tf.keras.layers.Dense(units = d_model,trainable = False)
        self.wk = tf.keras.layers.Dense(units = d_model,trainable = False)
        self.wv = tf.keras.layers.Dense(units = d_model,trainable = False)
        self.dense1 = tf.keras.layers.Dense(units = d_model,trainable = False)

    #x--input series divide into heads part, add the dimension of the turn of the head
    def split_into_heads(self,x,batch_size):

        x = tf.reshape(x,(batch_size,-1,self.num_heads,self.depth))

        return tf.transpose(x,perm = [0,2,1,3])

    def call(self,enc_output,dec_input,mask):
        batch_size = enc_output.shape[0]
        
        #linear layer 
        self.q = self.wq(dec_input)
        self.k = self.wk(enc_output)
        self.v = self.wq(enc_output)

        #splitting into heads
        q = self.split_into_heads(q,batch_size)
        k = self.split_into_heads(k,batch_size)
        v = self.split_into_heads(v,batch_size)

        attention_weights,multi_head_output = self_attention(q,k,v,mask,self.depth) #for every head

        pre_concat_multi_head_output =  tf.transpose(multi_head_output,perm = [0,2,1,3])
        post_concat_multi_head_output =  tf.reshape(pre_concat_multi_head_output,(batch_size,-1,self.d_model))
        
        final_output = self.dense1(post_concat_multi_head_output)

        return attention_weights,final_output
```

```python
#embedding --similarity, without the positional distance information. Thus add the position on embeddings
class get_positional_encoding(tf.keras.layers.Layer):
  def __init__(self,position,d_model):
    super(get_positional_encoding,self).__init__()
    self.position = position
    self.d_model = d_model
    self.pos_encoding = self.positional_encoding(position,d_model)
  
  def get_angle(self,position,i,d_model):
    angle = 1 / tf.pow(10000, (2*(i//2))/ tf.cast(d_model,tf.float32))
    return position * angle

  def positional_encoding(self, position, d_model): #d_model = 512 if 128
    angle_final = self.get_angle(tf.expand_dims(tf.range(tf.cast(position,tf.float32)),axis = 1),tf.expand_dims(tf.range(tf.cast(d_model,tf.float32)),axis = 0),d_model)
    np_angle_final = angle_final.numpy()
    np_angle_final[:,0::2] = np.sin(np_angle_final[:,0::2])
    np_angle_final[:,1::2] = np.cos(np_angle_final[:,1::2])
    angle_final = tf.convert_to_tensor(np_angle_final)
    pos_encoding = angle_final[tf.newaxis,...]
    return tf.cast(pos_encoding,dtype=tf.float32)
  
  def call(self,input):
    return input + self.pos_encoding[:, :tf.shape(input)[1], : ]
```

```python
#hidden_layer_shape--units(how many to enlarge) generally, d_model*4 
class FeedForward_Layer(tf.keras.layers.Layer): 
    def __init__(self,d_model,hidden_layer_shape):
        super(FeedForward_Layer,self).__init__()
        self.d_model = d_model
        self.hidden_layer_shape = hidden_layer_shape
        self.dense1 = tf.keras.layers.Dense(hidden_layer_shape, activation='relu') # (batch_size, seq_len, dff) relu-->gelu
        self.dense2 = tf.keras.layers.Dense(d_model)  # (batch_size, seq_len, d_model) normalization done on the final d_model axis

    def call(self,input):
        output = self.dense1(input)
        final_output = self.dense2(output)
        return final_output
  
```

```python
class Encoding_Layer(tf.keras.layers.Layer): #LN
    def __init__(self,num_heads,d_model,hidden_layer_shape,rate=0.1):
        super(Encoding_Layer,self).__init__()
        self.d_model = d_model
        self.num_heads = num_heads
        self.multi_head_attention = MultiHeadAttention(num_heads,d_model)
        self.ffd_net = FeedForward_Layer(d_model,hidden_layer_shape)
        self.norm1 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.norm2 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.dropout1 = tf.keras.layers.Dropout(rate)
        self.dropout2 = tf.keras.layers.Dropout(rate)
        norm_initializer = tf.keras.initializers.RandomNormal(mean=1.0, stddev=0.01)
#        self.norm1 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
#        self.norm2 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
    
    def call(self,input,mask,training = True):
        #part 1
        _,multihead_output = self.multi_head_attention(input,mask)
        dropout_multihead_output = self.dropout1(multihead_output,training = training)
        skip_dropout_output = tf.add(input,dropout_multihead_output)
        norm1_skip_output  = self.norm1(skip_dropout_output)        
        #part 2
        point_norm1_output = self.ffd_net(norm1_skip_output)
        dropout_point_output = self.dropout2(point_norm1_output,training = training)
        skip_dropout_output = tf.add(dropout_point_output,norm1_skip_output)
        norm2_skip_output = self.norm2(skip_dropout_output)
        return norm2_skip_output
```

```python
class Encoding_Layer(tf.keras.layers.Layer): #BN
    def __init__(self,num_heads,d_model,hidden_layer_shape,rate=0.1):
        super(Encoding_Layer,self).__init__()
        self.d_model = d_model
        self.num_heads = num_heads
        self.multi_head_attention = MultiHeadAttention(num_heads,d_model)
        self.ffd_net = FeedForward_Layer(d_model,hidden_layer_shape)
#        self.norm1 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
#        self.norm2 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.dropout1 = tf.keras.layers.Dropout(rate)
        self.dropout2 = tf.keras.layers.Dropout(rate)
        norm_initializer = tf.keras.initializers.RandomNormal(mean=1.0, stddev=0.01)
        self.norm1 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
        self.norm2 = layers.BatchNormalization(momentum=0.90, epsilon=10**-5, gamma_initializer=norm_initializer)
    
    def call(self,input,mask,training = True):
        #part 1
        _,multihead_output = self.multi_head_attention(input,mask)
        dropout_multihead_output = self.dropout1(multihead_output,training = training)
        skip_dropout_output = tf.add(input,dropout_multihead_output)
        norm1_skip_output  = self.norm1(skip_dropout_output)        
        #part 2
        point_norm1_output = self.ffd_net(norm1_skip_output)
        dropout_point_output = self.dropout2(point_norm1_output,training = training)
        skip_dropout_output = tf.add(dropout_point_output,norm1_skip_output)
        norm2_skip_output = self.norm2(skip_dropout_output)
        return norm2_skip_output
```

```python
class Decoding_Layer(tf.keras.layers.Layer):
    def __init__(self,num_heads,d_model,hidden_layer_shape,rate = 0.1):
        super(Decoding_Layer,self).__init__()
        self.multi_head_attention1 = MultiHeadAttention(num_heads,d_model)
        self.multi_head_attention2 = MultiHeadAttention2(num_heads,d_model)
        self.norm1 = tf.keras.layers.LayerNormalization(epsilon = 1e-6) #mean=1 variance=0
        self.norm2 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.norm3 = tf.keras.layers.LayerNormalization(epsilon = 1e-6)
        self.point_ffd_net = FeedForward_Layer(d_model,hidden_layer_shape)
        self.dropout1 = tf.keras.layers.Dropout(rate)
        self.dropout2 = tf.keras.layers.Dropout(rate)
        self.dropout3 = tf.keras.layers.Dropout(rate)

    def call(self,enc_output,dec_input,look_ahead_mask,padding_mask,training = True):
        #decoder attention with look ahead mask(include padding mask)
        # enc_output.shape --> (batch_size, input_seq_len, d_model)
        _,multihead1_output = self.multi_head_attention1(dec_input,look_ahead_mask)
        dropout1_multihead1_output = self.dropout1(multihead1_output)
        skip1_dropout1_output = tf.add(dropout1_multihead1_output,dec_input)
        norm1_skip1_output = self.norm1(skip1_dropout1_output)

        #decoder&encoder attention with padding mask
        _,multihead2_output  = self.multi_head_attention2(enc_output,dec_input,padding_mask)
        dropout2_multihead2_output = self.dropout2(multihead2_output)
        skip2_dropout2_output = tf.add(dropout2_multihead2_output,norm1_skip1_output)
        norm2_skip2_output = self.norm2(skip2_dropout2_output)

        #sum and normalize
        point_norm2_output = self.point_ffd_net(norm2_skip2_output)
        dropout3_point_output = self.dropout3(point_norm2_output)
        skip3_dropout3_output = tf.add(norm2_skip2_output,dropout3_point_output)
        norm3_skip3_output = self.norm3(skip3_dropout3_output)
        return norm3_skip3_output
```

```python
#without embeddings
class Encoder(tf.keras.Model):
    def __init__(self,d_model,num_heads,hidden_layer_shape,max_pos_encoding,num_enc_layers,rate = 0.1):
        super(Encoder,self).__init__()
        self.max_pos_encoding = max_pos_encoding #max_pos_encoding--longest position 
        self.num_enc_layers = num_enc_layers
        self.d_model = d_model
        self.num_heads = num_heads
        self.hidden_layer_shape = hidden_layer_shape
        self.position_encoding = get_positional_encoding(max_pos_encoding,d_model)#using part of the position
        self.encoder_layers = [Encoding_Layer(num_heads,d_model,hidden_layer_shape,rate) for _ in range(num_enc_layers)]
        self.dropout = tf.keras.layers.Dropout(rate)

    def call(self,inputs,training):
        enc_padding_mask = None
        pos_scaled_embedding = tf.math.add(inputs,self.position_encoding(inputs))
        output_input_encoder = self.dropout(pos_scaled_embedding,training = training)

        for i in range(self.num_enc_layers):
            output_input_encoder = self.encoder_layers[i](output_input_encoder,enc_padding_mask,training)

        return output_input_encoder
```

```python
#without embeddings
class Decoder(tf.keras.Model):
    def __init__(self,d_model,num_heads,hidden_layer_shape,max_position_encoding,num_layers,rate=0.1): #dec_input = decoder input
        super(Decoder,self).__init__()
        self.position_encoding = get_positional_encoding(max_position_encoding,d_model)
        self.d_model = d_model
        self.num_layers = num_layers
        self.decoder_layers = [Decoding_Layer(num_heads,d_model,hidden_layer_shape) for _ in tf.range(num_layers)]
        self.dropout = tf.keras.layers.Dropout(rate)
        self.num_layers = num_layers

    def call(self,input,training = True):
        #positional_encoding = self.position_encoding[:,:seq_len,:]
        dec_input=input[0]
        enc_output = input[1]
        look_ahead_mask = generate_lookahead_mask(dec_input)
#        dec_padding_mask = create_padding_mask(dec_input)
        padding_mask = None
        positional_encoding = self.position_encoding(dec_input)
        pos_scaled_embedding = tf.math.add(dec_input,positional_encoding)
        output_input_decoder = self.dropout(pos_scaled_embedding,training = training)

        for i in range(self.num_layers):
            output_input_decoder = self.decoder_layers[i](output_input_decoder,enc_output,look_ahead_mask,padding_mask)
        
        return output_input_decoder
```

```python
#With decoder
class Transformer(tf.keras.Model):
    def __init__(self,num_layers,num_heads,d_model,hidden_layer_shape,max_pos_encoding, rate = 0.1):
        super(Transformer,self).__init__()
        self.encoder = Encoder(d_model,num_heads,hidden_layer_shape,max_pos_encoding,num_layers,rate) #pe_input--max_pos_encoding
        self.decoder = Decoder(d_model,num_heads,hidden_layer_shape,max_pos_encoding,num_layers,rate) #pe_output--max_pos_encoding dec_input_Size--detect input size
        self.pooling = layers.AveragePooling1D(pool_size=10)
        self.final_layer = tf.keras.layers.Dense(1)

    def call(self,input,training = True):
        
        #mid = int(input.shape[0])
        enc_input=input[0]
        dec_input=input[1]
        enc_output = self.encoder(enc_input,training)
#        final_dec_input = tf.concat(dec_input,enc_output,0)
        final_dec_input = [dec_input,enc_output]
        dec_output = self.pooling(self.decoder(final_dec_input,training))
        final_output = self.final_layer(dec_output)

        return final_output
```

```python
#single class accuracy calculation
#positive-->1 class negtive-->0 class
def cal_acc(result,threshold):
    batch_size = result.shape[0]//2 
    result = np.sum(np.squeeze(result, axis=2),axis=1)
    result=np.where(result<threshold,0,1)
    accuracy = (np.sum(result[:batch_size]==1)+np.sum(result[batch_size:]==0))/len(result)*100 #neg_batch label: 0 && pos_batch label: 1
    return accuracy
```

```python
#single class loss calculation
def single_loss(result):
    batch_size = result.shape[0]//2
    sum1 = 0.0
    sum2 = 0.0 
    loss = 0.0 
    l_sum = [] 
    for i in range(batch_size):
        y_pos_max = tf.math.reduce_max(result[i])
        y_neg_max = tf.math.reduce_max(result[i+batch_size])
        l = tf.math.reduce_max([0, tf.math.add(1.0, tf.math.subtract(y_neg_max, y_pos_max))]) #波形1セット分の損失
        l_sum.append(l)

    loss = tf.math.reduce_sum(l_sum)

    pos_result = np.squeeze(result[0:batch_size], axis=2).flatten()
    neg_result = np.squeeze(result[batch_size:], axis=2).flatten()
    number = []
    for k in range(len(pos_result)-1):
        number.append(tf.square(tf.math.subtract(pos_result[k], pos_result[k+1])))
    sum1 = tf.math.reduce_sum(number)
    sum2 = tf.math.reduce_sum(pos_result)
    smooth = tf.math.scalar_mul(m1, sum1)
    sparse = tf.math.scalar_mul(m2, sum2)
    lasso = tf.math.add(smooth, sparse)
    loss = tf.math.add(loss, lasso)
    return loss
```

```python
#single class loss calculation
def multi_loss(result):
    batch_size = result.shape[0]//2
    sum1 = 0.0
    sum2 = 0.0 
    loss = 0.0 
    l_sum = [] 
    for i in range(batch_size):
        y_pos_max = tf.math.reduce_max(result[i])
        y_neg_max = tf.math.reduce_max(result[i+batch_size])
        l = tf.math.reduce_max([0, tf.math.add(1.0, tf.math.subtract(y_neg_max, y_pos_max))]) #波形1セット分の損失
        l_sum.append(l)

    loss = tf.math.reduce_sum(l_sum)

    pos_result = np.squeeze(result[0:batch_size], axis=2).flatten()
    neg_result = np.squeeze(result[batch_size:], axis=2).flatten()
    number = []
    for k in range(len(pos_result)-1):
        number.append(tf.square(tf.math.subtract(pos_result[k], pos_result[k+1])))
    sum1 = tf.math.reduce_sum(number)
    sum2 = tf.math.reduce_sum(pos_result)
    smooth = tf.math.scalar_mul(m1, sum1)
    sparse = tf.math.scalar_mul(m2, sum2)
    lasso = tf.math.add(smooth, sparse)
    loss = tf.math.add(loss, lasso)
    return loss

```

```python
#Cross Validation
def cross_division(start,data,batch_size,split):
  ''' start: refers to the Kth Cross-validation data
      batch_size: batch_size 
      split: K
  '''
  val_start = start * batch_size // split  #val_start: the real start dividing point in the original datasets
  val_end = (start +1) * batch_size // split #val_end: the real end dividing point in the original datasets
  train_batch = np.vstack((data[:val_start],data[val_end:]))
  test_batch = data[val_start:val_end]
  return train_batch,test_batch
  
```

```python
#training batch = 200 cross validation transformer BN
max_step = 200 # 学習の繰り返し回数 every step will train K times for K-Fold validation
m1 = 0.001 #正規化項内の定数
m2 = 0.001
loss_value =[] #train loss list
acc_train = [] #train accuracy list
pos_data,neg_data = CWRUCVdataset.processed_dataset_ab,CWRUCVdataset.processed_dataset #positive data and negative data
threshold = 0.6 #threshold for model to assume whether the batch is abnormal or not
val_loss_value = [] #validation loss list
val_acc_value = [] #validation accuracy list
test_batch_size = batch_size//split*2 #positive training batch + negative training batch
train_batch_size = batch_size*2 - test_batch_size #positive test batch + negative test batch
for step in tqdm(range(max_step)):
  if step == max_step:
    break
  val_loss = [] #validation loss in every K epochs validatipon 
  val_acc = [] #validation accuracy in every K epochs validatipon
  #Cross-validation training and validation 
  for val_start in list(range(0,split)): #val_star: validation set divide start index
    pos_train_batch,pos_test_batch = cross_division(val_start,pos_data,batch_size,split) #divide for the validation set
    neg_train_batch,neg_test_batch = cross_division(val_start,neg_data,batch_size,split)
    train_batch = tf.concat([pos_train_batch,neg_train_batch],0)
    test_batch = tf.concat([pos_test_batch,neg_test_batch],0)
    loss = 0.0
    #traing dataset for training 
    with tf.GradientTape() as t:
      result = tf.zeros([train_batch_size, 0, 1]) 
      result = tf.cast(result,tf.float32)
  
      for i in range(0, 200, 10):
        buf = model(train_batch[:, i:i+10])
        result = tf.concat([result, buf], axis=1)
      loss = single_loss(result) #calculate the training loss
      loss_value.append(loss) #save in the training loss list
      train_acc = cal_acc(result,threshold) #calculate the training accuracy
      acc_train.append(train_acc) #save in the training accuracy list

    d = t.gradient(loss, model.trainable_weights)
    optimizer.apply_gradients(zip(d, model.trainable_weights))
    print('Train loss: =',loss,'Train acc: =',train_acc)

    #validation dataset only for testing
    test_result = tf.zeros([test_batch_size, 0, 1])
    for i in range(0, 200, 10):
        test_buf = model(test_batch[:, i:i+10])
        test_result = tf.concat([test_result, test_buf], axis=1)
    test_loss = single_loss(test_result) #calculate for every K epoch testing loss
    test_acc = cal_acc(test_result,threshold) #save in the every K epoch testing loss list
    val_loss.append(test_loss)#calculate for testing accuracy
    val_acc.append(test_acc) #save in the testing accuracy list
  val_final_loss = np.mean(val_loss) #for every K-fold, calculate the mean loss as the final loss
  val_final_acc = np.mean(val_acc) #for every K-fold, calculate the mean accuracy as the final loss
  print('Val Loss: ', val_final_loss,'Val Acc: ',val_final_acc)
  val_loss_value.append(val_final_loss) #save in the final testing loss list
  val_acc_value.append(val_final_acc)#save in the accuracy testing loss list
```
